```python
import matplotlib.pyplot as plt

import utils
import preprocessing
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = preprocessing.load_dataset()
```

```python
outliers = []
for i in np.linspace(0.01, 0.3, 30):
    clf = IsolationForest(contamination=i)
    clf.fit(df)
    y_pred = clf.predict(df)
    outliers.append((i, sum(y_pred == -1)))
```

```python
plt.scatter(x = [i for i in np.linspace(0.01, 0.3, 30)] ,
            y = outliers)
```
